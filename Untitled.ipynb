```python
def uva(data,variable):
    plt.figure(figsize = (10,len(variable)*5))
    for j,i  in enumerate(variable):
        mini = data[i].min()
        maxi = data[i].max()
        mean = data[i].mean()
        median = data[i].median()
        st_dev = data[i].std()
        skew = data[i].skew()
        kurt = data[i].kurtosis()
        points = mean-st_dev, mean+st_dev
        
        size = len(variable)
        plt.subplot(len(variable),1,j+1)
        plt.title(i)
        sns.kdeplot(data[i],fill = True)
        sns.lineplot(x=points, y=[0,0], color = 'black', label = "std_dev")
        plt.scatter(x = mean,y = 0, color = 'red',label = 'mean')
        plt.scatter(x = median,y = 0, color = 'green',label = 'median')
        plt.legend()
        
     
        plt.title(' \n\n kurtosis = {}; skew = {};mean = {}; median = {}'.format(round(kurt,2),round(skew,2),round(mean,2),round(median,2)))
```

### Missing Value Treatement

```python
data.isna().sum()
```

```python
# Imputing missing values
data.MinorAxisLength = data.MinorAxisLength.fillna(data.MinorAxisLength.median())
data.ConvexArea = data.ConvexArea.fillna(data.ConvexArea.median())
data.Compactness = data.Compactness.fillna(data.Compactness.median())
data.ShapeFactor1 = data.ShapeFactor1.fillna(data.ShapeFactor1.median())
data.ShapeFactor4 = data.ShapeFactor4.fillna(data.ShapeFactor4.median())
```

```python
data.isna().sum()
```

### Outlier Treatement

```python
# There are a lot of outliers in the data as the difference between maximum value and the 3rd quartile of a variable is large
data.describe()
```

```python
columns = numerical.columns
for i in range(0,len(columns)):
    Q1 = np.quantile(data[columns[i]],0.25)
    Q3 = np.quantile(data[columns[i]],0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    data[columns[i]].loc[data[columns[i]] < lower_bound] = lower_bound
    data[columns[i]].loc[data[columns[i]] > upper_bound] = upper_bound
```

```python
# As it is seen, the difference between 3rd Quartile and the maximum value has been reduced
data.describe()
```

# Modeling

```python
x = data.drop(['Class'],axis = 1)

# Label encoding the dependent variable
data['Class'] = data['Class'].map({'DERMASON': 1, 'SIRA': 2, 'SEKER':3,'HOROZ':4,'CALI':5,'BARBUNYA':6,'BOMBAY':7})

y = data.Class
```

```python
# Scaling and Train Test splitting

scaler = MinMaxScaler()
x = scaler.fit_transform(x)
x = pd.DataFrame(x,columns = numerical.columns)
train_x,test_x,train_y,test_y = tts(x,y,random_state = 56)
```

#### Ridge regularisation

```python
# Ridge logistic reression model
def ridge(data,train_x,train_y,test_x):
    alpha = 1
    ridge_model = Ridge(alpha=alpha)
    ridge_model.fit(train_x, train_y)
    ridge_train = ridge_model.predict(train_x)
    ridge_test = ridge_model.predict(test_x)
# r2 score
    return (r2_score(test_y, ridge_test).round(2))
```

```python
ridge(data,train_x,train_y,test_x)
```

#### Models with Hyperparameter tuning

```python
def model(model,train_x,train_y,test_x):
    model.fit(train_x,train_y)
    pred_train = model.predict(train_x)
    pred_test = model.predict(test_x)
    return (r2_score(model.predict(test_x),test_y).round(2))
```

```python
# Decision Tree with Random search Hyperparameter tuning
dt = DecisionTreeClassifier (random_state = 10)
parameter_grid = {
    'max_depth' : [4,5,6,7,8,9],
    'max_features': [0.3,0.5, 0.7,0.9]
    }
randomsearch_dt = RandomizedSearchCV(estimator=dt, param_distributions=parameter_grid, n_iter= 10, cv=5)
randomsearch_dt.fit(train_x,train_y)
max_feat_dt = randomsearch_dt.best_params_['max_features']
max_dep_dt =  randomsearch_dt.best_params_['max_depth']
dt = DecisionTreeClassifier (max_features = max_feat_dt, max_depth = max_dep_dt,random_state = 10) 

# Random Forest Classifier with Random search Hyperparameter tuning

clf = RandomForestClassifier(random_state = 96)
parameter_grid = { 
'n_estimators':[2,3,4,5,6],
'max_depth':[2,3,4,5,6]}
randomsearch_rf = RandomizedSearchCV(estimator=clf, param_distributions=parameter_grid, n_iter= 10, cv=5)
randomsearch_rf.fit(train_x,train_y)
n_est_rf= randomsearch_rf.best_params_['n_estimators']
max_dep_rf = randomsearch_rf.best_params_['max_depth']
clf =  RandomForestClassifier(n_estimators = n_est_rf , max_depth = max_dep_rf,random_state = 96) 

# Adaptive Boosting with Random search Hyperparameter tuning

ada = AdaBoostClassifier()
parameter_grid = { 
'n_estimators':[2,3,4,5],
'learning_rate':[2.0,3.0,4.0,5.0]}
randomsearch_ada = RandomizedSearchCV(estimator=ada, param_distributions=parameter_grid, n_iter= 10, cv=5)
randomsearch_ada.fit(train_x,train_y)
n_estimators_ada = randomsearch_ada.best_params_['n_estimators']
learning_rate_ada = randomsearch_ada.best_params_['learning_rate']
ada = AdaBoostClassifier(n_estimators = n_estimators_ada,learning_rate = learning_rate_ada, random_state=96)
```

```python
# Decision tree model with hyperparameter tuning
model(dt,train_x,train_y,test_x)
```

```python
# Random forest model with hyperparameter tuning
model(clf,train_x,train_y,test_x)
```

```python
# Gradient boosting model with hyperparameter tuning
model(ada,train_x,train_y,test_x)
```

## Stacking
#### These models whose hyperparameters have been tuned are then stacked

```python
def model_stacking(model,train_x,train_y,test_x):
    model.fit(train_x,train_y)
    pred_train = model.predict(train_x)
    pred_test = model.predict(test_x)
    return pred_train,pred_test
```

```python
dt_train,dt_test = model_stacking(dt,train_x,train_y,test_x)
```

```python
clf_train,clf_test = model_stacking(clf,train_x,train_y,test_x)
```

```python
ada_train,ada_test = model_stacking(ada,train_x,train_y,test_x)
```

```python
train_predictions = {"M1":dt_train,
                    "M2":clf_train,
                    "M3":ada_train}
train_predictions = pd.DataFrame(train_predictions)
train_predictions.head(2)
```

```python
test_predictions = {"M1":dt_test,
                   "M2":clf_test,
                   "M3":ada_test}
test_predictions = pd.DataFrame(test_predictions)
test_predictions.head(2)
```
